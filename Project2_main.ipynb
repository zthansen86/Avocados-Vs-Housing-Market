```python
top_city_data3 = top_city_data2.rename(columns={"Atlanta, GA":"Atlanta",
                                                "Chicago, IL":"Chicago",
                                               "Austin, TX": "Austin", 
                                                "Miami-Fort Lauderdale, FL": "Miami", 
                                                "New York, NY": "NewYork",
                                                "Los Angeles-Long Beach-Anaheim, CA": "LosAngeles",
                                                "Portland, OR": "Portland", 
                                                "Seattle, WA": "Seattle", 
                                                "Nashville, TN": "Nashville"})

top_city_data3                                               
```

```python
avocado_4=pd.DataFrame(avocado_3)
avocado_4
```

```python
avocado_5 = avocado_4.groupby(['region'])
avocado_5.head()
```

```python
combined_data = pd.merge(avocado_4, top_city_data3, on =["Date"], how='outer') 
combined_data
```

```python
combined_data2=combined_data.dropna(how='any')
combined_data2
```

```python
cities = ['Atlanta']
Atlanta_Data = combined_data2[combined_data2.region.isin(cities) == True]
Atlanta_Data1 = Atlanta_Data.drop(columns=["region", "Total Bags", "Small Bags", "Large Bags", "XLarge Bags", "type", "Chicago", "Austin", "Miami", "NewYork", "LosAngeles", "Portland", "Seattle", "Nashville"]).reset_index()  
Atlanta_Data1 = Atlanta_Data1.rename(columns={"AveragePrice": "ATL_AveragePrice",
                                             "Total Volume": "ATL_Total Volume"})
                                             
Atlanta_Data1                                                 
                                                                                               
```

```python
cities = ['Chicago']
Chicago_Data = combined_data2[combined_data2.region.isin(cities) == True]
Chicago_Data1 = Chicago_Data.drop(columns=["region", "Total Bags", "Small Bags", "Large Bags", "XLarge Bags", "type", "Atlanta", "Austin", "Miami", "NewYork", "LosAngeles", "Portland", "Seattle", "Nashville"]).reset_index()  
Chicago_Data1 = Chicago_Data1.rename(columns={"AveragePrice": "CHI_AveragePrice",
                                             "Total Volume": "CHI_Total Volume"})
                                            
Chicago_Data1
```

```python
cities = ['NewYork']
NewYork_Data = combined_data2[combined_data2.region.isin(cities) == True]
NewYork_Data1 = NewYork_Data.drop(columns=["region", "Total Bags", "Small Bags", "Large Bags", "XLarge Bags", "type", "Atlanta", "Chicago", "Austin", "Miami", "LosAngeles", "Portland", "Seattle", "Nashville"]).reset_index() 
NewYork_Data1 = NewYork_Data1.rename(columns={"AveragePrice": "NY_AveragePrice",
                                             "Total Volume": "NY_Total Volume"})
                                             
NewYork_Data1
```

```python
cities = ['Portland']
Portland_Data = combined_data2[combined_data2.region.isin(cities) == True]
Portland_Data1 = Portland_Data.drop(columns=["region", "Total Bags", "Small Bags", "Large Bags", "XLarge Bags","type", "Atlanta", "Chicago", "Austin", "Miami", "LosAngeles", "NewYork", "Seattle", "Nashville"]).reset_index()
Portland_Data1 = Portland_Data1.rename(columns={"AveragePrice": "PTL_AveragePrice",
                                             "Total Volume": "PTL_Total Volume"})
                                             
Portland_Data1
```

```python
cities = ['Seattle']
Seattle_Data = combined_data2[combined_data2.region.isin(cities) == True]
Seattle_Data1 = Seattle_Data.drop(columns=["region", "Total Bags", "Small Bags", "Large Bags", "XLarge Bags","type", "Atlanta", "Chicago", "Austin", "Miami", "LosAngeles", "NewYork", "Portland", "Nashville"]).reset_index()
Seattle_Data1 = Seattle_Data1.rename(columns={"AveragePrice": "SEA_AveragePrice",
                                             "Total Volume": "SEA_Total Volume"})
                                           
Seattle_Data1
```

```python
cities = ['Nashville']
Nashville_Data = combined_data2[combined_data2.region.isin(cities) == True]
Nashville_Data1 = Nashville_Data.drop(columns=["region", "Total Bags", "Small Bags", "Large Bags", "XLarge Bags","type", "Atlanta", "Chicago", "Austin", "Miami", "LosAngeles", "NewYork", "Portland", "Seattle"]).reset_index()
Nashville_Data1 = Nashville_Data1.rename(columns={"AveragePrice": "NSH_AveragePrice",
                                             "Total Volume": "NSH_Total Volume"})
                                             
Nashville_Data1
```

```python
cities = ['LosAngeles']
LosAngeles_Data = combined_data2[combined_data2.region.isin(cities) == True]
LosAngeles_Data1 = LosAngeles_Data.drop(columns=["region", "Total Bags", "Small Bags", "Large Bags", "Nashville", "XLarge Bags","type", "Atlanta", "Chicago", "Austin", "Miami", "NewYork", "Portland", "Seattle"]).reset_index()
LosAngeles_Data1 = LosAngeles_Data1.rename(columns={"AveragePrice": "LA_AveragePrice",
                                             "Total Volume": "LA_Total Volume"})
                                             
LosAngeles_Data2 = LosAngeles_Data1.set_index("Date").drop(columns=["index"])
LosAngeles_Data2
```

```python
merge1 = pd.merge(Atlanta_Data1, Chicago_Data1, on=["Date"], how="outer").set_index("Date").drop(columns=["index_x", "index_y"])

merge1
```

```python
merge2 = pd.merge(NewYork_Data1, Portland_Data1, on=["Date"], how="outer").set_index("Date").drop(columns=["index_x", "index_y"])

merge2
```

```python
merge3 = pd.merge(Seattle_Data1, Nashville_Data1, on=["Date"], how="outer").set_index("Date").drop(columns=["index_x", "index_y"])

merge3
```

```python
merge4 = pd.merge(LosAngeles_Data2, merge1, on=["Date"], how="outer")

merge4
```

```python
merge5 = pd.merge(merge2, merge4, on=["Date"], how="outer")

merge5
```

```python
merge_main = pd.merge(merge3, merge5, on=["Date"], how="outer")

merge_main
```
